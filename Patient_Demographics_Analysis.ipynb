# Patient Demographics Analysis
This notebook presents a basic analysis of patient demographics, focusing on age, gender, and race distributions within the diabetic patient dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetic_data.csv')
```

## Distribution of Patients by Age

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age', order=sorted(df['age'].unique()))
plt.title('Distribution of Patients by Age')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Distribution of Patients by Gender

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gender')
plt.title('Distribution of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## Distribution of Patients by Race

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='race', order=df['race'].value_counts().index)
plt.title('Distribution of Patients by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
